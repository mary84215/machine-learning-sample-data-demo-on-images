```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import scipy.io
```

```python
from sklearn.cluster import KMeans
import seaborn as sns 
```

```python
from skimage import io
```

```python
image = io.imread('11HW2_PatchPanelsGrey.jpeg')
#io.imshow(image)
#io.show()
```

```python
rows = image.shape[0]
cols = image.shape[1]
```

```python
image = image.reshape(image.shape[0]*image.shape[1],3)
```

```python
plt.figure(figsize=(20,4))
k=[2,5,10,15,20]
for i in range(5):
    km = KMeans(n_clusters = k[i])
    km.fit(image)
    plt.subplot(1,5,i+1) 
    io.imshow(km.predict(image).reshape((rows,cols)),cmap="gray")
```

```python
image = io.imread('11HW2_PatchPanels.jpg')
#io.imshow(image)
#io.show()
rows = image.shape[0]
cols = image.shape[1]
image = image.reshape(image.shape[0]*image.shape[1],3)
plt.figure(figsize=(20,4))
k=[2,5,10,15,20]
for i in range(5):
    km = KMeans(n_clusters = k[i])
    km.fit(image)
    plt.subplot(1,5,i+1) 
    io.imshow(km.predict(image).reshape((rows,cols)))
```
